# Makes change - All solutions recursive with best solution

```python
import time

# Function to count occurrences of items in a list
def count_occurrences(array):
    counts = {}
    for item in array:
        if item in counts:
            counts[item] += 1  # If the item is already in the dictionary, increment its count
        else:
            counts[item] = 1  # If the item is not in the dictionary, add it with a count of 1
    return counts

# Global variable to count the number of combinations
combination_count = 0

# Variables to store the best combination and its coin count
best_combination = []
best_coin_count = float('inf')

# Function to calculate all combinations of coins to make a specific amount
def calculate_change_combinations(coins, amount, combination=[], start=0):
    global combination_count, best_combination, best_coin_count  # Use the global variables
    tolerance = 1e-9  # A small tolerance value for floating-point comparisons

    # Check if amount is within tolerance of 0
    if abs(amount) < tolerance:
        combination_count += 1  # Increment the combination count

        # Count the occurrences of coins used to make the target amount
        counts = count_occurrences(combination)
        coin_count = sum(counts.values())  # Calculate the total coin count

        # Check if this combination uses fewer coins than the current best
        if coin_count < best_coin_count:
            best_combination = combination.copy()  # Update the best combination
            best_coin_count = coin_count  # Update the best coin count
    elif amount < 0 or start >= len(coins):
        return
    else:
        for i in range(start, len(coins)):
            coin = coins[i]
            combination.append(coin)  # Add the current coin to the combination
            calculate_change_combinations(coins, round(amount - coin, 2), combination, i)  # Recursively calculate combinations
            combination.pop()  # Remove the current coin from the combination after exploring possibilities
```

```python
# Global variable to count the number of combinations
combination_count = 0

# Variables to store the best combination and its coin count
best_combination = []
best_coin_count = float('inf')

# Define the list of coins and the amount of change
coin_list = [5, 2, 1, 0.5, 0.2, 0.1, 0.05]
print("The types of coins we have are:", coin_list)
change_amount = 12.35
print("The amount of change we need to return is:", change_amount)

start_time = time.time()  # Record the start time
# Call the function to calculate and display the combinations of coins for the given amount
calculate_change_combinations(coin_list, change_amount)
end_time = time.time()  # Record the end time
execution_time = end_time - start_time

# Print the best combination
print("Best Combination:")
print("+-------+-----------+")
print("| Count |    Coin   |")
print("+-------+-----------+")

counts = count_occurrences(best_combination)
sorted_counts = sorted(counts.items(), key=lambda x: x[0], reverse=True)

for coin, coin_count in sorted_counts:
    print(f"|   {coin_count}   |   {coin:.2f} €  |")

print("+-------+-----------+\n")

# Print the total number of coins used by the best solution
print(f"Total number of coins used by the best solution: {best_coin_count}")

# Verify if the total value of coins used in the best solution equals the original change amount
best_combination_total = round(sum(best_combination), 2)
if best_combination_total == change_amount:
    print("Verification: Best solution's change is correct.")
else:
    print("Verification: Best solution's change is incorrect.")
    missing_or_extra = change_amount - best_combination_total
    if missing_or_extra > 0:
        print(f"Missing change: {missing_or_extra:.2f} euros")
    elif missing_or_extra < 0:
        print(f"Extra change: {abs(missing_or_extra):.2f} euros")

# Print the total number of combinations
print(f"Total number of combinations: {combination_count}")

print(f"Execution time: {execution_time:.6f} seconds")
```
