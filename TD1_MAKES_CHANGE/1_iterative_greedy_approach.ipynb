# Makes change - Iterative greedy approach

```python
import time

def calculate_change_combinations(coins, amount):
    cents_amount = int(amount * 100)  # Convert euros to cents
    i = 0  # Initialize the index to 0
    total_coins = 0  # Initialize the total number of coins
    change_used = 0  # Initialize the total value of coins used

    print(f"Change for {amount:.2f} euros:")
    print("+-------+-----------+")
    print("| Count |    Coin   |")
    print("+-------+-----------+")

    while cents_amount != 0 and i < len(coins):
        coin_value_cents = int(coins[i] * 100)  # Convert coin value to cents

        # Calculate the number of coins of the current denomination
        coin_count = cents_amount // coin_value_cents

        if coin_count > 0:
            print(f"|   {coin_count}   |   {coins[i]:.2f} €  |")

        # Update the remaining amount, total number of coins, and total value of coins used
        cents_amount %= coin_value_cents
        total_coins += coin_count
        change_used += coin_count * coin_value_cents

        # Move to the next denomination
        i += 1

    print("+-------+---------+")
    print(f"Total number of coins: {total_coins}")

    # Verify if the total value of coins used equals the original amount
    if change_used == int(amount * 100):
        print("Verification: Change is correct.")
    else:
        print("Verification: Change is incorrect.")
        missing_or_extra = int(amount * 100) - change_used
        if missing_or_extra > 0:
            print(f"Missing change: {missing_or_extra / 100:.2f} euros")
        elif missing_or_extra < 0:
            print(f"Extra change: {abs(missing_or_extra) / 100:.2f} euros")
```

```python
# Define the list of coins and the amount of change
coin_list = [5, 2, 1, 0.5, 0.2, 0.1, 0.05]
print("The types of coins we have are:", coin_list)
change_amount = 12.35
print("The amount of change we need to return is:", change_amount)

start_time = time.time()  # Record the start time
# Call the function with the inputs
calculate_change_combinations(coin_list, change_amount)
end_time = time.time()  # Record the end time
execution_time = end_time - start_time
print(f"Execution time: {execution_time:.6f} seconds")
```
