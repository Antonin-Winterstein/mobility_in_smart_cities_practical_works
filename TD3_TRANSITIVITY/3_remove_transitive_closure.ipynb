# Transitivity - Remove transitive closure to a matrix

```python
def remove_transitive_closure_to_matrix(matrix):
  # Reflexive reduction
  for i in range(len(matrix)):
      matrix[i][i] = 0  # Set diagonal elements to 0 (reflexive reduction)

  # Transitive reduction
  for j in range(len(matrix)):  # Iterate through columns (j)
      for i in range(len(matrix)):  # Iterate through rows (i)
          if matrix[i][j]:  # Check if there is a direct link from i to j
              for k in range(len(matrix)):  # Iterate through columns again (k)
                  if matrix[j][k]:  # Check if there is a direct link from j to k
                      matrix[i][k] = 0  # Set the transitive link from i to k to 0
  return matrix
```

```python
matrix = [
    [1, 1, 1],
    [0, 1, 1],
    [0, 0, 1]]

print("Original Matrix:")
for row in matrix:
    print(row)

# Print the transitive closure matrix
print("Remove Transitive Closure Matrix:")
for row in remove_transitive_closure_to_matrix(matrix):
    print(row)
```
