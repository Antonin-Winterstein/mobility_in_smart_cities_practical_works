# Transitivity - Add transitive closure to a matrix

```python
def add_transitive_closure_to_matrix(matrix):
  # Iterate through the matrix to compute the transitive closure
  for i in range(len(matrix)):  # Loop through rows
      for j in range(len(matrix)):  # Loop through columns
          # Check if there is a direct link from j to i
          if matrix[j][i]:
              for k in range(len(matrix)):  # Loop through columns again
                  # Check if there is a direct link from i to k
                  if matrix[i][k]:
                      # Update the matrix to indicate a transitive link from j to k
                      matrix[j][k] = 1

  return matrix
```

```python
matrix = [
    [1, 1, 1],
    [0, 1, 1],
    [0, 0, 1]]

print("Original Matrix:")
for row in matrix:
    print(row)

# Print the transitive closure matrix
print("Add Transitive Closure Matrix:")
for row in add_transitive_closure_to_matrix(matrix):
    print(row)
```
